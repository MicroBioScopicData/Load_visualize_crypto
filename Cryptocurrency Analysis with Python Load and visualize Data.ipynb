## Loading data from yfinance
In this section, we will use yfinance to retrieve cryptocurrency data and store it in a pandas DataFrame. yfinance is a popular library for downloading financial data directly from Yahoo Finance. It provides easy access to stock, index, ETF, and of course cryptocurrency data.


```python
# Install the yfinance library by running the following command in your Python environment:
# !pip install yfinance

# Import the yfinance library
import yfinance as yf


# Define the start and end dates for the data
start = '2021-01-01'
end = '2023-03-29'

# Define the list of cryptocurrencies to download
cryptos = ['BTC-USD', 'ETH-USD', 'XRP-USD', 'LTC-USD', 'BCH-USD']

# Use yfinance to download the data for each cryptocurrency 
df_cryptos = yf.download(cryptos, start=start, end=end)

# Select only the close column for every cryptocurrency
df = df_cryptos["Close"]

df.describe()
df.info()
df
```

### Candlestick charts
Candlestick charts are often used to analyze short-term price movements in financial markets. They are particularly useful for technical analysis and identifying patterns such as support and resistance levels. Candlestick charts provide more information than line charts as they display the open, high, low, and close prices for each time period, making it easier to identify patterns such as trends, reversal signals, and gaps. They are also useful for visualizing price volatility and trading ranges. To create a candlestick chart of the BTC, we use the mpf.plot function from the mplfinance library.


```python
import yfinance as yf
# Install mplfinance
# !pip install mplfinance
import pandas as pd
import mplfinance as mpf

# Get Bitcoin data from Yahoo Finance
df = yf.download(tickers='BTC-USD', period='1mo', interval='1d')
df
# Create candlestick chart
mpf.plot(df, type='candle',  figratio=(12,8), title='BTC-USD',savefig='Candlestick_3.png')
print("Chart saved as Candlestick_3.png")
```

### To visualize the trading volume of the Bitcoin 
Analyzing the trading volume of a cryptocurrency can provide valuable insights into the market activity and liquidity of the cryptocurrency. For example, if we observe a significant increase in trading volume over a particular period, it could indicate a surge in demand or interest in cryptocurrency, potentially leading to an increase in its price. On the other hand, a decrease in trading volume could signal a lack of market interest or activity, which could lead to a decrease in cryptocurrency price. Additionally, analyzing the trading volume of cryptocurrency can also help identify market trends and patterns, such as the occurrence of high trading volume during certain times of the day or week. To visualize the trading volume of the cryptocurrency (BTC), create a bar chart using the plot.bar() function. For example, to create a bar chart of Bitcoin's trading volume over the last month, run the following:


```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Get Bitcoin data from Yahoo Finance
df = yf.download(tickers='BTC-USD', period='2y', interval='1d')

# Plot the volume as a bar chart
plt.figure(figsize=(12, 8))
plt.bar(df.index, df['Volume'], width=0.5, color='gray')
plt.title('Bitcoin Volume')
plt.xlabel('Date')
plt.ylabel('Volume (BTC)')
plt.tight_layout()
plt.savefig("bar_chart.jpg")
```

```python
# Get Bitcoin data from Yahoo Finance
df = yf.download(tickers='BTC-USD', period='2y', interval='1d')

# Set up the figure and subplots using gridspec
fig = plt.figure(figsize=(10, 8))
gs = fig.add_gridspec(nrows=5, ncols=1, height_ratios=[3, 1, 1, 1, 1])

# Plot the historical stock prices on the top subplot
ax1 = fig.add_subplot(gs[:3, :])
ax1.plot(df.index, df['Close'])
ax1.set_title('Historical stock prices of BTC-USD')

# Plot the trading volume on the bottom subplot
ax2 = fig.add_subplot(gs[3:, :])
ax2.bar(df.index, df['Volume'])
ax2.set_title('BTC-USD Volume')
plt.tight_layout()
plt.savefig("bar_chart_line.jpg")
```
